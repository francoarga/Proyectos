#### Analisis de clientes 

```python
df['Age'].fillna(value = df['Age'].mean)
```

```python
df['Age'] = df['Age'].astype(int) 
df.loc[df['Age'] < 30, 'Grupo por edad'] = 'Joven'
df.loc[(df['Age'] >= 30) & (df['Age'] < 50), 'Grupo por edad'] = 'Adulto'
df.loc[df['Age'] > 50, 'Grupo por edad'] = 'Senior'
df 
```

```python
df['AverageSpend'] = df['AverageSpend'].astype(float)
mujeres = df['Gender'] == 'Female'
mujer = df[mujeres]
mujer_joven = mujer['Grupo por edad'] == 'Joven'
mujer_adulta = mujer['Grupo por edad'] == 'Adulto'
mujer_señora = mujer['Grupo por edad'] == 'Senior'
```

```python
mj = mujer[mujer_joven]
mj_mean = mj['AverageSpend'].mean()

ma = mujer[mujer_adulta]
ma_mean = ma['AverageSpend'].mean()

ms = mujer[mujer_señora]
ms_mean = ms['AverageSpend'].mean()
```

```python
hombres = df['Gender'] == 'Male'
hombre = df[hombres]
hombre_joven = hombre['Grupo por edad'] == 'Joven'
hombre_adulto = hombre['Grupo por edad'] == 'Adulto'
hombre_señor = hombre['Grupo por edad'] == 'Senior'
```

```python
hj = hombre[hombre_joven]
hj_mean = hj['AverageSpend'].mean()

ha = hombre[hombre_adulto]
ha_mean = ha['AverageSpend'].mean()

hs = hombre[hombre_señor]
hs_mean = hs['AverageSpend'].mean()
```

```python
df.loc[(df['Age'] < 30) & (df['Gender'] == 'Male'), 'Gasto por edad'] = hj_mean
df.loc[(df['Age'] >= 30) & (df['Age'] < 50) & (df['Gender'] == 'Male'), 'Gasto por edad'] = ha_mean
df.loc[(df['Age'] > 50) & (df['Gender'] == 'Male'), 'Gasto por edad'] = hs_mean
df.loc[(df['Age'] < 30) & (df['Gender'] == 'Female'), 'Gasto por edad'] = mj_mean
df.loc[(df['Age'] >= 30) & (df['Age'] < 50) & (df['Gender'] == 'Female'), 'Gasto por edad'] = ma_mean
df.loc[(df['Age'] > 50) & (df['Gender'] == 'Female'), 'Gasto por edad'] = ms_mean
df
 
```

```python
import matplotlib.pyplot as plt
import numpy as np

data_main = {
    'grupo_edad': ['Joven', 'Joven', 'Adulto', 'Adulto', 'Senior', 'Senior'],
    'genero': ['Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre', 'Mujer'],
    'monto_total': [hj_mean, mj_mean, ha_mean, ma_mean, hs_mean, ms_mean]
}

gastos_totales = pd.DataFrame(data_main)

fig, ax = plt.subplots()

gastos_totales.pivot(index='grupo_edad', columns='genero', values='monto_total').plot(kind='bar', ax=ax)

ax.set_title('Gasto promedio por edad y genero')
ax.set_xlabel('Grupo por edad')
ax.set_ylabel('Gasto promedio')

# Ajustar las marcas del eje y para tener más números
max_value = gastos_totales['monto_total'].max()
step = 5  # Define el paso entre las marcas
yticks = np.arange(0, max_value + step, step)
ax.set_yticks(yticks)
ax.set_yticklabels(yticks)

ax.tick_params(axis='y', labelsize=8)

plt.show()
```

#### Analisis de transacciones 

```python
df_transacciones = pd.read_csv('C:/Users/franc/Documents/Retail_Transactions_Dataset.csv')
transacciones = pd.DataFrame(df_transacciones)
transacciones
```

```python
df_transacciones['Date'] = pd.to_datetime(df_transacciones['Date'])
```

```python
df_transacciones['Mes y año'] = df_transacciones['Date'].dt.to_period('M')
```

```python
monto_por_mes = df_transacciones.groupby('Mes y año')['Total_Cost'].sum().reset_index()
df_monto = pd.DataFrame(monto_por_mes)
df_monto
```

```python
# 5. Crear un gráfico de líneas
plt.figure(figsize=(18, 6))
plt.plot(df_monto['Mes y año'].astype(str), df_monto['Total_Cost'], marker='o')
plt.title('Evolución Mensual del Monto Total de las Transacciones')
plt.xlabel('Mes')
plt.ylabel('Monto Total')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
