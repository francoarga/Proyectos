# Ejercicio 1a)

**Mostrar que el vector de unos es autovector de las matrices $L$ y $R$.  
¿Qué autovalor tiene? ¿Qué agrupación representa?**

### Laplaciano $L$


El laplaciano de un grafo se define como:

$L = K - A$

donde:
- $K$ es la matriz diagonal de grados, con $K_{ii} = k_i$
- $A$ es la matriz de adyacencia del grafo

Sea el vector de unos:

$1 = \begin{pmatrix} 1 \\ 1 \\ . \\ . \\ .  \\ 1 \end{pmatrix}$

Multiplicamos:

$L . 1 = (K - A) . 1 = K . 1 - A . 1$

Pero:

- $K . 1 = \begin{pmatrix} k_1 \\ k_2 \\ . \\ . \\ .  \\ k_n \end{pmatrix}$  
  
  . 

- $A . 1 = \begin{pmatrix} \sum_j A_{1j} \\ \sum_j A_{2j} \\ . \\ . \\ .  \\ \sum_j A_{nj} \end{pmatrix} = \begin{pmatrix} k_1 \\ k_2 \\ . \\ . \\ .  \\ k_n \end{pmatrix}$

Entonces:

$L . 1 = k - k = 0$

Por lo tanto, $1$ es **autovector de $L$** con **autovalor 0**.

### Matriz de modularidad R

La matriz de modularidad está definida como:

$R = A - P$

donde:

$P_{ij} = \frac{k_i k_j}{2E}$

Si multiplicamos:

$R . 1 = A . 1 - P . 1$

Ya sabemos que $A . 1 = k$.  
Ahora calculemos $P . 1$ para la fila $i$:

$$
(P . 1)_i = \sum_j P_{ij} = \sum_j \frac{k_i k_j}{2E} = \frac{k_i}{2E} \sum_j k_j = \frac{k_i}{2E} . 2E = k_i
$$

Entonces:

$$
P . 1 = k
\Rightarrow R . 1 = k - k = 0
$$

Por lo tanto, $1$ también es **autovector de $R$** con **autovalor 0**.

#### Interpretación

El vector $1$ representa la agrupación **trivial**, en la que **todos los nodos pertenecen a una sola comunidad**.

Esto tiene sentido:
- En el caso del laplaciano, no hay corte si todos los nodos están juntos.
- En el caso de la modularidad, no hay ganancia sobre el modelo nulo si no se hace ninguna partición.

#### Conclusión 
El vector de unos es autovector de ambas matrices $L$ y $R$, y el autovalor asociado en ambos casos es $0$. Representa la partición trivial donde toda la red es una sola comunidad.

# Ejercicio 1b)

**Mostrar que si $L$ o $R$ tienen dos autovectores $v_1$ y $v_2$ asociados a autovalores distintos $\lambda_1 \ne \lambda_2$, entonces $v_1^T v_2 = 0$**

#### Fundamento teórico

Sea $M$ una matriz simétrica real (como es el caso de $L$ y $R$), y supongamos que $v_1$ y $v_2$ son autovectores de $M$ asociados a autovalores $\lambda_1$ y $\lambda_2$ respectivamente, con $\lambda_1 \ne \lambda_2$.

Queremos demostrar que $v_1^T v_2 = 0$, es decir, que $v_1$ y $v_2$ son ortogonales.

#### Demostración

Partimos de la ecuación de autovalores:

$$
M v_1 = \lambda_1 v_1 \quad \text{y} \quad M v_2 = \lambda_2 v_2
$$

Tomamos el producto escalar $v_1^T M v_2$.

Por un lado, usando que $v_2$ es autovector:

$$
v_1^T M v_2 = v_1^T (\lambda_2 v_2) = \lambda_2 v_1^T v_2
$$

Por otro lado, usamos que $M$ es simétrica:

$$
v_1^T M v_2 = (M v_1)^T v_2 = (\lambda_1 v_1)^T v_2 = \lambda_1 v_1^T v_2
$$

Entonces:

$$
\lambda_1 v_1^T v_2 = \lambda_2 v_1^T v_2
\Rightarrow (\lambda_1 - \lambda_2) v_1^T v_2 = 0
$$

Y como $\lambda_1 \ne \lambda_2$, se concluye:

$$
v_1^T v_2 = 0
$$

#### Conclusión

Si $v_1$ y $v_2$ son autovectores de una **matriz simétrica real** como $L$ o $R$, y están asociados a **autovalores distintos**, entonces **son ortogonales**:

$$
v_1^T v_2 = 0
$$

Esta propiedad es muy importante en métodos espectrales, ya que permite trabajar con bases ortogonales de autovectores.

# Ejercicio 1c)


**Mostrar que si $v$ es un autovector de autovalor distinto de cero de $L$ o $R$, entonces se cumple:**

$$
\sum_i v_i = 0
$$

#### Fundamento teórico

Sea $M$ una matriz simétrica como $L$ o $R$, y sea $v$ un autovector de $M$ con autovalor $\lambda \ne 0$.  
Recordemos que tanto $L$ como $R$ son **simétricas** y que el **vector de unos** $1$ es también autovector, con autovalor **cero**:

$$
M . 1 = 0
$$

#### Demostración

Tomamos el producto escalar entre $v$ y el vector $1$:

$$
1^T M v = (M . 1)^T v = 0^T v = 0
$$

Pero también, como $v$ es autovector:

$$
1^T M v = 1^T (\lambda v) = \lambda . (1^T v)
$$

Entonces:

$$
\lambda . \sum_i v_i = 0 \quad \Rightarrow \quad \sum_i v_i = 0
$$


#### Conclusión

Todo autovector $v$ de $L$ o $R$ que tenga **autovalor distinto de cero** cumple que **la suma de sus componentes es cero**.  
Esto significa que está ortogonal al vector $1$, es decir:

$$
v \perp 1
$$

#### Shifting de autovalores

Sea $M \in \mathbb{R}^{n \times n}$ una matriz diagonalizable, con autovalores $\lambda_1, \lambda_2, \ldots, \lambda_n$, y autovectores $v_1, v_2, \ldots, v_n$ asociados respectivamente.

Queremos demostrar que:

1. Los autovalores de la matriz $M + \mu I$ son $\gamma_i = \lambda_i + \mu$

2. Los autovectores de $M + \mu I$ son los mismos que los de $M$

3. Si $\lambda_i + \mu \neq 0$ para todo $i$, entonces $M + \mu I$ es inversible

#### Demostración

Sea $v_i$ un autovector de $M$ asociado al autovalor $\lambda_i$:

$$
Mv_i = \lambda_i v_i
$$

Entonces:

$$
(M + \mu I)v_i = Mv_i + \mu v_i = (\lambda_i + \mu)v_i \Rightarrow \gamma_i = \lambda_i + \mu
$$

Los autovectores no cambian.

Si $\lambda_i + \mu \neq 0$ para todo $i$, entonces todos los autovalores de $M + \mu I$ son distintos de cero, lo que implica que la matriz es inversible.

#### Conclusión

- Los autovalores de $M + \mu I$ son $\lambda_i + \mu$
- Los autovectores son los mismos
- Si ninguno de los $\lambda_i + \mu$ es cero, la matriz es inversible

# Ejercicio 2b)

### Método de la potencia inverso

Sea $L$ una matriz simétrica semidefinida positiva (por ejemplo, un laplaciano), y sea $\mu > 0$.

#### 1. Invertibilidad

Todos los autovalores de $L$ cumplen $\lambda_i \geq 0$.  
Entonces $L + \mu I$ tiene autovalores $\lambda_i + \mu > 0$, y por lo tanto es inversible.


#### 2. Aplicar la potencia inversa

Sabemos que:

$$
Lv = \lambda v \Rightarrow (L + \mu I)v = (\lambda + \mu)v \Rightarrow (L + \mu I)^{-1}v = \frac{1}{\lambda + \mu}v
$$


El método de la potencia aplicado a $(L + \mu I)^{-1}$ devuelve el autovector asociado al mayor valor de $\frac{1}{\lambda + \mu}$, que corresponde al menor $\lambda$.

#### 3. Caso particular

Para el laplaciano de un grafo conexo:

- El menor autovalor es 0
- El autovector asociado es el vector 1
- Entonces $(L + \mu I)^{-1}$ tiene autovalor $\frac{1}{\mu}$ asociado a 1

### Deflacion de Hotelling


Queremos mostrar que la matriz:

$$
\tilde{M} = M - \frac{\lambda_1}{v_1^T v_1} v_1 v_1^T
$$

Tiene los mismos autovectores que $M$, y que el autovalor asociado a $v_1$ se hace 0.

#### Demostración

Si $M v_1 = \lambda_1 v_1$, entonces:

$$
\tilde{M} v_1 = M v_1 - \frac{\lambda_1}{v_1^T v_1} v_1 (v_1^T v_1) = \lambda_1 v_1 - \lambda_1 v_1 = 0
$$

Ahora si $v_k$ es otro autovector ortogonal a $v_1$, se cumple:

$$
\tilde{M} v_k = \left( M - \frac{\lambda_1}{v_1^T v_1} v_1 v_1^T \right) v_k = M v_k - \frac{\lambda_1}{v_1^T v_1} v_1 (v_1^T v_k)
$$

Como $v_1^T v_k = 0$, el segundo término se anula:

$$
\tilde{M} v_k = M v_k = \lambda_k v_k
$$

#### Conclusión

- $\tilde{M}$ tiene los mismos autovectores que $M$
- El autovalor de $v_1$ pasa a ser 0
- El resto de los autovalores no cambia

Esto permite anular el dominante y encontrar el segundo mayor aplicando el método de la potencia nuevamente.

### Comparación y conclusiones

Ambos métodos permiten identificar comunidades en la red, pero priorizan objetivos distintos: el método del **Laplaciano** busca minimizar la cantidad de aristas que conectan diferentes grupos, mientras que la **modularidad** evalúa si la densidad de conexiones internas dentro de las comunidades supera la esperada en un grafo aleatorio con los mismos grados.

En este caso, el método del Laplaciano logró una partición en cuatro comunidades con un mínimo de **8 aristas cortadas**, aunque la partición óptima posible en 4 grupos cortaría solo **6 aristas**, lo que indica que el método no siempre encuentra la mejor solución. Esto ocurre porque el algoritmo depende del **segundo autovector del Laplaciano** para decidir cómo partir cada subgrafo, y cuando los **autovalores son cercanos**, el autovector correspondiente puede variar numéricamente entre ejecuciones. Esto genera que, ante pequeñas fluctuaciones, se produzcan **particiones distintas**, lo cual explica por qué no siempre converge a la partición óptima incluso con el mismo valor de `niveles`.

Además, como el método es recursivo, un error temprano (una mala partición en los primeros niveles) puede arrastrarse y empeorar la calidad final del corte. Esto también se evidenció al correr el método varias veces y observar diferentes cantidades de aristas cortadas.

En contraste, el método de la **modularidad** produjo una partición clara y estable en dos comunidades ([0, 1, 2, 3] y [4, 5, 6, 7]), con solo **2 aristas cortadas**, maximizando el valor de Q. Como tiene un **criterio interno de parada** (sólo se subdivide si mejora la modularidad global), resulta más robusto ante fluctuaciones numéricas, y tiende a capturar mejor la estructura general del grafo.

En resumen, aunque ambos enfoques son útiles y complementarios, **en este caso el método de la modularidad resultó más efectivo**, al encontrar una partición más coherente y con menor pérdida de conexiones internas que el método basado en el Laplaciano.

### Aplicación a la red real de museos

En esta sección se aplican los métodos de detección de comunidades sobre la red real de museos de la Ciudad de Buenos Aires, construida en el TP1 a partir de sus coordenadas geográficas. Se utilizaron los algoritmos basados en Laplaciano y en modularidad, evaluando su desempeño sobre versiones simetrizadas de la matriz de adyacencia para distintos valores de `m` (cantidad de vecinos conectados). El objetivo es observar cómo varía la estructura de comunidades detectada en función de `m`, comparando el número, tamaño y distribución espacial de las comunidades generadas por cada método.


```python
# coordenadas
museos = gpd.read_file('https://raw.githubusercontent.com/MuseosAbiertos/Leaflet-museums-OpenStreetMap/refs/heads/principal/data/export.geojson')
barrios = gpd.read_file('https://cdn.buenosaires.gob.ar/datosabiertos/datasets/ministerio-de-educacion/barrios/barrios.geojson')
museos = museos[~museos.geometry.is_empty].copy()
museos = museos.to_crs("EPSG:3857")
museos['x'] = museos.geometry.x
museos['y'] = museos.geometry.y
```

```python
# Calcular matriz de distancias
coords = np.array([[geom.x, geom.y] for geom in museos.geometry])
D = pairwise_distances(coords, metric='euclidean')
```

```python
# Función para simetrizar la matriz de adyacencia
def simetriza_adyacencia(A):
    return np.ceil((A + A.T) / 2).astype(int)

# Función para graficar comunidades con mapa de fondo
def plot_comunidades(df_museos, etiquetas, titulo, barrios):
    fig, ax = plt.subplots(figsize=(10, 8))
    barrios.to_crs("EPSG:3857").plot(ax=ax, facecolor='white', edgecolor='black', alpha=0.5)
    scatter = ax.scatter(df_museos['x'], df_museos['y'], c=etiquetas, cmap='tab20', s=50)
    ax.set_title(titulo)
    plt.colorbar(scatter, ax=ax, label="Comunidad")
    plt.axis('off')
    plt.show()

# Análisis para diferentes valores de m
valores_m = [3, 5, 10, 50]
matrices_adyacencia = {}

for m in valores_m:
    print(f"\n>>> Análisis para m = {m}")
    
    # Construir matriz de adyacencia simétrica
    A = construye_adyacencia(D, m)
    A_sym = simetriza_adyacencia(A)
    matrices_adyacencia[m] = A_sym

    # ------------------------
    # MÉTODO DE MODULARIDAD
    # ------------------------
    comunidades_mod = TP2_template_funciones.modularidad_iterativo(A=A_sym)
    etiquetas_mod = np.zeros(A_sym.shape[0], dtype=int)
    for i, grupo in enumerate(comunidades_mod):
        for nodo in grupo:
            etiquetas_mod[nodo] = i
    n_mod = len(comunidades_mod)
    print(f"  Modularidad → {n_mod} comunidades")
    plot_comunidades(museos, etiquetas_mod, f"Modularidad (m={m})", barrios)

    # ------------------------
    # MÉTODO DEL LAPLACIANO
    # ------------------------
    niveles = int(np.log2(n_mod)) if n_mod > 1 else 0
    comunidades_lap = TP2_template_funciones.laplaciano_iterativo(A_sym, niveles=niveles)
    etiquetas_lap = np.zeros(A_sym.shape[0], dtype=int)
    for i, grupo in enumerate(comunidades_lap):
        for nodo in grupo:
            etiquetas_lap[nodo] = i
    print(f"  Laplaciano → {len(comunidades_lap)} comunidades (niveles = {niveles})")
    plot_comunidades(museos, etiquetas_lap, f"Laplaciano (m={m})", barrios)
```

**Análisis del impacto del parámetro *m* sobre la detección de comunidades**

Al aumentar el valor de *m*, es decir, la cantidad de vecinos más cercanos con los que se conecta cada museo en la red, observamos una tendencia clara: **disminuye la cantidad de comunidades detectadas** tanto por el método basado en **modularidad** como por el basado en el **laplaciano espectral**. En todos los casos, el número de comunidades halladas por modularidad fue ligeramente mayor, lo que es consistente con su objetivo explícito de maximizar la diferenciación entre subgrupos conectados.

Este comportamiento puede explicarse por el efecto estructural que produce un mayor *m*: **la red se vuelve progresivamente más densa**, ya que cada nodo incorpora más conexiones. Esto genera una topología más homogénea y menos modular, dificultando la identificación de regiones bien diferenciadas. Como consecuencia, las divisiones naturales del grafo \(presentes en valores bajos de *m*\) se desdibujan, y los algoritmos tienden a agrupar un mayor número de nodos dentro de comunidades más grandes y difusas.

En valores bajos como \( *m* = 3 \), tanto la modularidad como el laplaciano detectan múltiples comunidades pequeñas, muchas de las cuales presentan una fuerte coherencia espacial: tienden a concentrarse en zonas geográficas bien delimitadas. Sin embargo, a medida que *m* aumenta (por ejemplo, a \( *m* = 10 \) o \( *m* = 50 \)), se observa una clara consolidación de las comunidades, donde los nodos previamente separados por la escasa conectividad ahora forman parte de grupos comunes.

Este fenómeno alcanza su punto crítico cuando *m* es suficientemente grande como para que la red se aproxime a una estructura **casi completamente conexa**. En ese caso, las diferencias topológicas entre nodos se diluyen, y los algoritmos dejan de detectar cortes significativos. La red pasa de tener múltiples agrupamientos locales a mostrar una estructura global más uniforme, en la que las particiones son menos informativas.

En términos prácticos, este análisis destaca la importancia de **ajustar correctamente el valor de m** según el nivel de granularidad deseado. Un *m* muy bajo puede fragmentar en exceso la red, mientras que un *m* muy alto puede ocultar subestructuras relevantes. La selección adecuada del parámetro es clave para obtener una segmentación útil e interpretable de la red de museos.

### Conclusión general

A lo largo de los dos trabajos prácticos se abordó un análisis integral de la red de museos de la Ciudad de Buenos Aires, utilizando dos aproximaciones complementarias: el modelado dinámico mediante caminatas aleatorias (TP1) y la detección de comunidades a partir de métodos espectrales (TP2). La articulación entre ambos enfoques permitió construir una visión profunda y multiescalar del comportamiento de la red, con implicancias tanto teóricas como aplicadas.

En el TP1, se utilizó el algoritmo de **PageRank** para estimar la centralidad de los museos, entendida como la probabilidad de ser alcanzado tras múltiples pasos en la red. Se observó que, con valores bajos del parámetro de amortiguación α (por ejemplo, α = 1/7), el ranking reflejaba con mayor fidelidad la estructura topológica: los museos más conectados o ubicados en zonas densas tendían a obtener mayores puntajes. En cambio, con α altos (como 6/7), la aleatoriedad dominaba y las diferencias entre nodos se atenuaban. Además, al variar el parámetro **m** (cantidad de vecinos), se notó que en redes más locales (`m=1` o `m=3`) algunos museos periféricos alcanzaban altos valores de PageRank, mientras que en redes más densas (`m=10` o `m=50`), los museos céntricos pasaban a ocupar las posiciones dominantes.

En el TP2, se aplicaron técnicas espectrales basadas en los autovalores y autovectores de las matrices **Laplaciana** y de **modularidad**. En particular, al aplicar ambos métodos sobre un grafo de ejemplo en el punto 3, se observó que la modularidad fue más efectiva, al lograr una partición más estable y con menor cantidad de aristas cortadas, mientras que el método del laplaciano mostró sensibilidad a pequeñas variaciones y no siempre alcanzó el corte óptimo. Ambas matrices permitieron transformar problemas de segmentación (minimizar corte o maximizar modularidad) en problemas algebraicos, resolubles mediante el método de la potencia y sus variantes (potencia inversa, deflación de Hotelling). Las particiones obtenidas mostraron agrupamientos geográficos coherentes: barrios del centro tendían a formar comunidades densas, mientras que en la periferia aparecían subestructuras más aisladas. Se observó además que al aumentar el valor de `m`, ambos métodos detectaban menos comunidades, pero más grandes, reflejando una transición hacia una red más densamente conectada y homogénea.

Una comparación directa entre **PageRank y clustering** reveló coincidencias y contrastes: algunos nodos con alto PageRank coincidieron con hubs dentro de las comunidades detectadas, pero otros funcionaron como “puentes” entre comunidades distintas, destacando por su conectividad global aunque no quedaran atrapados dentro de un único grupo estructural. Esto refuerza la idea de que ambos métodos miden propiedades distintas: PageRank evalúa influencia o atracción en dinámicas de navegación, mientras que el clustering revela cohesión interna en la estructura de enlaces.

Cabe destacar que el algoritmo de PageRank también puede interpretarse como un **problema espectral**: el vector de ranking es el **autovector principal** (asociado al autovalor 1) de una matriz estocástica modificada por el factor de amortiguación. Esta interpretación refuerza la unidad conceptual entre los métodos vistos en ambos TPs: todos se reducen, en última instancia, a encontrar autovectores de matrices derivadas de la red, resolviendo así problemas de optimización sobre la estructura.

Desde el punto de vista metodológico, ambos TPs permitieron aplicar herramientas de álgebra lineal computacional a grafos reales, abordando problemas complejos de forma eficiente. La reutilización del **método de la potencia** sobre distintas matrices (Laplaciana y modularidad en TP2, matriz de transición en TP1) mostró la versatilidad de estas técnicas y ayudó a interiorizar los fundamentos teóricos detrás de ellas.

En síntesis, este recorrido dejó en claro que analizar redes no es una tarea unidimensional. Las nociones de centralidad y comunidad son complementarias, y su combinación permite enriquecer la interpretación. Mientras el TP1 enfatizó el flujo de visitantes y la importancia individual de los museos, el TP2 aportó una segmentación estructural que puede ser útil, por ejemplo, para diseñar rutas, clasificar museos por zona de influencia o entender cómo se organizan espontáneamente los trayectos de los asistentes. Esta sinergia entre dinámica y estructura habilita futuras aplicaciones, como combinar ambos métodos para sugerir visitas dentro de una comunidad relevante, priorizando nodos centrales en cada grupo.
