```python
# Devolver una lista con los código y nombres de departamentos, asociados al nombre de la provincia al que pertenecen.
provincia_depto = """
                SELECT departamento.id,
                departamento.descripcion AS departamento,
                provincia.descripcion AS provincia
                FROM departamento
                INNER JOIN provincia
                ON id_provincia = provincia.id;
              """ 
provincia_deptoSQL = sql^ provincia_depto 
print(provincia_deptoSQL) 
```

```python
# Devolver los casos registrados en la provincia de “Chaco”.
consultaSQL = """
                SELECT *
                FROM provincia_deptoSQL
                INNER JOIN casos
                ON  provincia_deptoSQL.id = casos.id_depto
                WHERE provincia = 'Chaco';
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Devolver aquellos casos de la provincia de “Buenos Aires” cuyo campo cantidad supere los 10 casos.
Buenos_Aires = """
                SELECT *
                FROM provincia_deptoSQL
                INNER JOIN casos
                ON  provincia_deptoSQL.id = casos.id_depto
                WHERE provincia = 'Buenos Aires';
              """ 
Buenos_AiresSQL = sql^ Buenos_Aires 

mas_de_diez = """
                SELECT *
                FROM Buenos_AiresSQL
                WHERE cantidad > 10.0;
              """ 
mas_de_diezSQL = sql^ mas_de_diez
mas_de_diezSQL
```

```python
# Devolver un listado con los nombres de los departamentos que no tienen ningún caso asociado.
consultaSQL = """
                SELECT departamento.*
                FROM departamento
                LEFT OUTER JOIN casos
                ON  departamento.id = casos.id_depto
                WHERE casos.id_depto IS NULL;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Devolver un listado con los tipos de evento que no tienen ningún caso asociado.
consultaSQL = """
                SELECT tipoevento.*
                FROM tipoevento
                LEFT OUTER JOIN casos
                ON  tipoevento.id = casos.id_tipoevento
                WHERE casos.id_depto IS NULL;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Calcular la cantidad total de casos que hay en la tabla casos.
consultaSQL = """
                SELECT count(cantidad) FROM casos;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Calcular la cantidad total de casos que hay en la tabla casos para cada año y cada tipo de caso. Presentar la información de la siguiente manera: descripción del tipo de caso, año y cantidad. Ordenarlo por tipo de caso (ascendente) y año (ascendente).
consultaSQL = """
                SELECT id_tipoevento, anio, count(*) AS cantidad,
                FROM casos
                GROUP BY anio, id_tipoevento
                ORDER BY anio ASC, id_tipoevento ASC;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Misma consulta que el ítem anterior, pero sólo para el año 2019.
consultaSQL = """
                SELECT id_tipoevento, anio, count(*) AS cantidad,
                FROM casos
                WHERE anio = 2019
                GROUP BY anio, id_tipoevento
                ORDER BY anio ASC, id_tipoevento ASC;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Calcular la cantidad total de departamentos que hay por provincia. Presentar la información ordenada por código de provincia.
consultaSQL = """
                SELECT id_provincia, count(*) AS cantidad_depto_por_provincia,
                FROM departamento
                GROUP BY id_provincia
                ORDER BY id_provincia ASC;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Listar los departamentos con menos cantidad de casos en el año 2019.
consultaSQL = """
                SELECT MIN(cantidad) menos_casos
                FROM departamento
                INNER JOIN casos
                ON departamento.id = casos.id_depto
                WHERE anio = 2019
                GROUP BY descripcion;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Listar los departamentos con más cantidad de casos en el año 2020.
consultaSQL = """
                SELECT MAX(cantidad) mas_casos
                FROM departamento
                INNER JOIN casos
                ON departamento.id = casos.id_depto
                WHERE anio = 2020
                GROUP BY descripcion
                ORDER BY mas_casos DESC
                LIMIT 5;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Listar el promedio de cantidad de casos por provincia y año.
consultaSQL = """
                SELECT id_provincia, anio, AVG(cantidad) promedio,
                FROM departamento
                INNER JOIN casos
                ON departamento.id = casos.id_depto
                GROUP BY id_provincia, anio;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Listar, para cada provincia y año, cuáles fueron los departamentos que más cantidad de casos tuvieron.
consultaSQL = """
                SELECT descripcion, MAX(cantidad) maximo,
                FROM departamento
                INNER JOIN casos
                ON departamento.id = casos.id_depto
                GROUP BY id_provincia, anio, descripcion
                ORDER BY maximo DESC
                LIMIT 15;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Mostrar la cantidad de casos total, máxima, mínima y promedio que tuvo la provincia de Buenos Aires en el año 2019.
provinciaSQL = """
                SELECT * 
                FROM departamento
                INNER JOIN provincia
                ON departamento.id_provincia = provincia.id;
              """ 
provincias = sql^ provinciaSQL 
consultaSQL = """
                SELECT MIN(cantidad) AS minimo, MAX(cantidad) AS maximo, SUM(cantidad) AS total 
                FROM provincias
                INNER JOIN casos
                ON provincias.id = casos.id_depto
                WHERE descripcion_1 = 'Buenos Aires' AND anio = 2019;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Misma consulta que el ítem anterior, pero sólo para aquellos casos en que la cantidad total es mayor a 1000 casos.
consultaSQL = """
                SELECT descripcion_1 AS provincia, MIN(cantidad) AS minimo, MAX(cantidad) AS maximo, SUM(cantidad) AS total 
                FROM provincias
                INNER JOIN casos
                ON provincias.id = casos.id_depto
                GROUP BY anio, id_provincia, descripcion_1;
              """ 
dataframeResultado = sql^ consultaSQL 
consultaSQL2 = """
                SELECT *
                FROM dataframeResultado
                WHERE total > 1000;
              """ 
dataframeResultado2 = sql^ consultaSQL2 
dataframeResultado2
```

```python
# Listar los nombres de departamento (y nombre de provincia) que tienen mediciones tanto para el año 2019 como para el año 2020. Para cada uno de ellos devolver la cantidad de casos promedio. Ordenar por nombre de provincia (ascendente) y luego por nombre de departamento (ascendente).
consultaSQL = """
                SELECT descripcion_1 AS provincia, descripcion AS departamento, AVG(cantidad) as PROMEDIO
                FROM provincias
                INNER JOIN casos
                ON provincias.id = casos.id_depto
                WHERE cantidad > 0 AND anio = 2019 OR anio = 2020
                GROUP BY anio, id_provincia, descripcion_1, descripcion
                ORDER BY descripcion_1 ASC, descripcion ASC;
              """ 
dataframeResultado = sql^ consultaSQL 
dataframeResultado
```

```python
# Devolver una tabla que tenga los siguientes campos: descripción de tipo de evento, id_depto, nombre de departamento, id_provincia, nombre de provincia, total de casos 2019, total de casos 2020.
consultaSQL = """
                SELECT cantidad AS total_de_casos_2019, id_depto, anio
                FROM casos
                WHERE anio = 2019;
              """ 
dataframeResultado = sql^ consultaSQL
consultaSQL2 = """
                SELECT cantidad AS total_de_casos_2020, id_depto, anio, id_tipoevento,
                FROM casos
                WHERE anio = 2020;
              """ 
dataframeResultado2 = sql^ consultaSQL2
consultaSQL3 = """
                SELECT total_de_casos_2019, total_de_casos_2020, dataframeResultado.id_depto, id_tipoevento,
                FROM dataframeResultado
                INNER JOIN dataframeResultado2
                ON dataframeResultado.id_depto = dataframeResultado2.id_depto;
              """ 
dataframeResultado3 = sql^ consultaSQL3
consultaSQL4 = """
                SELECT total_de_casos_2019, total_de_casos_2020, id_depto, descripcion AS descripcion_de_tipo_evento,
                FROM tipoevento
                INNER JOIN dataframeResultado3
                ON tipoevento.id = dataframeResultado3.id_tipoevento;
              """ 
dataframeResultado4 = sql^ consultaSQL4 
consultaSQL5 = """
                SELECT descripcion_de_tipo_evento, id_depto, descripcion AS nombre_de_departamento, id_provincia, descripcion_1 AS nombre_de_provincia, SUM(total_de_casos_2019) AS total_de_casos_2019, SUM(total_de_casos_2020) AS total_de_casos_2020,
                FROM dataframeResultado4
                INNER JOIN provincias
                ON dataframeResultado4.id_depto = provincias.id
                GROUP BY descripcion_de_tipo_evento, id_depto, descripcion, id_provincia, descripcion_1;
              """ 
dataframeResultado5 = sql^ consultaSQL5 
dataframeResultado5

```
